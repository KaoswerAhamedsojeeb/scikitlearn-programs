```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
ir_data = load_iris()
```

```python
dir(ir_data)
```

```python
df = pd.DataFrame(ir_data["data"], columns=ir_data["feature_names"])
```

```python
df["target"] = ir_data["target"]
```

```python
df["target_name"] = [ir_data.target_names[i] for i in ir_data.target]
```

```python
df
```

```python
df= df.drop(['sepal length (cm)','sepal width (cm)'], axis=1)
```

```python
df
```

```python
sns.scatterplot(data= df, x='petal length (cm)', y= 'petal width (cm)', hue='target_name')
```

```python
scaler = MinMaxScaler()
```

```python
df[["petal length (cm)","petal width (cm)"]]=scaler.fit_transform(df[["petal length (cm)","petal width (cm)"]])
```

```python
df
```

```python
sns.scatterplot(data= df, x="petal length (cm)", y="petal width (cm)", hue='target_name')
plt.show()
```

```python
df=df.drop(["target_name"], axis=1)
```

```python
df
```

```python
inertia = [] 

K = range(1, 11)  

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(K, inertia, 'bo-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia (SSE)')
plt.title('Elbow Method for Optimal k')
plt.grid(True)
plt.show()
```

```python
 k_means = KMeans(n_clusters=3, random_state=42)
```

```python
predicted_val = k_means.fit_predict(df)
predicted_val
```

```python
df["cluster"]= predicted_val
```

```python
df
```

```python
sns.scatterplot(data=df, x='petal length (cm)', y='petal width (cm)', hue ="cluster")
plt.show()
```
