```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv('basic_logistic_insurance.csv')
```

```python
df.head()
```

```python
sns.scatterplot(data = df, x='age', y='have_insurance')
plt.show()
```

```python
x_train,x_test, y_train,y_test = train_test_split(df[['age']], df.have_insurance, test_size=0.2)
```

```python
len(x_train)
```

```python
log_reg_model = LogisticRegression()
```

```python
log_reg_model.fit(x_train,y_train)
```

```python
log_reg_model.predict(pd.DataFrame([{"age":45}]))
```

```python
log_reg_model.score(x_test,y_test)
```
