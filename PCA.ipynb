# PCA -> principal component analysis

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
```

```python
digit = load_digits()
```

```python
dir(digit)
```

```python
df= pd.DataFrame(digit.data)
df
```

```python
df.describe()
```

```python
scaled_data = StandardScaler().fit_transform(df)
scaled_data
```

```python
x_train, x_test, y_train, y_test = train_test_split(scaled_data, digit.target, test_size=0.2, random_state=42)
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
pca_data = PCA(0.90).fit_transform(scaled_data)
```

```python
pca_data.shape
```

```python
x_train_pca, x_test_pca, y_train_pca, y_test_pca = train_test_split(pca_data, digit.target, test_size=0.2, random_state=42)
```

```python
model2 = LogisticRegression()
model2.fit(x_train_pca,y_train_pca)
```

```python
model2.score( x_test_pca,y_test_pca)
```
