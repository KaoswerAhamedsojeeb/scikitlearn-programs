```python
import pandas as pd
from sklearn.datasets import load_iris
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

```python
data = load_iris()
```

```python
dir(data)
```

```python
data['feature_names']
```

```python
data['data']
```

```python
df = pd.DataFrame(data['data'], columns=data['feature_names'])
```

```python
df['target'] = data.target
```

```python
df.head()
```

```python
data['target_names']
```

```python
df['flower_name'] = df.target.apply(lambda x:data.target_names[x])
```

```python
df.head()
```

```python
sns.scatterplot(data=df, x='sepal length (cm)', y='sepal width (cm)', hue="flower_name", marker="D")
plt.show()
```

```python
sns.scatterplot(data=df, x='petal length (cm)', y='petal width (cm)', hue="flower_name")
plt.show()
```

```python
x=df.drop(['target','flower_name'], axis=1)
y=df['target']
```

```python
x
```

```python
y
```

```python
x_train,x_test, y_train,y_test = train_test_split(x,y, test_size=0.2)
```

```python
len(x_train)
```

```python
model = SVC()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
