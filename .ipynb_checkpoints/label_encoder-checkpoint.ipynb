```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd
```

```python
df = pd.read_csv("Housing.csv")
```

```python
df.head()
```

```python
new_df=df
le_mainroad = LabelEncoder()
le_guestroom = LabelEncoder()
le_basement	 = LabelEncoder()
le_hotwaterheating = LabelEncoder()
le_airconditioning = LabelEncoder()
le_prefarea=LabelEncoder()
```

```python
new_df['mainroad_n'] = le_mainroad.fit_transform(new_df['mainroad'])
new_df['guestroom_n'] = le_guestroom.fit_transform(new_df['guestroom'])
new_df['basement_n'] = le_basement.fit_transform(new_df['basement'])
new_df['hotwaterheating_n'] = le_hotwaterheating.fit_transform(new_df['hotwaterheating'])
new_df['airconditioning_n'] = le_airconditioning.fit_transform(new_df['airconditioning'])
new_df['prefarea_n'] = le_prefarea.fit_transform(new_df['prefarea'])
```

```python
new_df=new_df.drop(['mainroad','guestroom','basement','hotwaterheating','airconditioning','prefarea'],axis=1)
```

```python
new_df
```

```python
dum = pd.get_dummies(new_df['furnishingstatus']).astype(int)
```

```python
dum
```

```python
new_df=new_df.drop('furnishingstatus', axis=1)
```

```python
new_df= pd.concat([new_df,dum],axis=1)
```
